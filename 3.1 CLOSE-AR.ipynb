#### RMSE value indicates the average magnitude of the errors

#### MAPE value indicates the average percentage difference between the predicted and actual values.

```python
def rmsemape(y_Test,predicted_stock_price_test_ori):
    from sklearn.metrics import mean_squared_error
    print("RMSE-Testset:",mean_squared_error(y_Test,predicted_stock_price_test_ori,squared=False))
    #print("RMSE-Trainset:",mean_squared_error(y_Train,predicted_stock_price_train_ori,squared=False))
    from sklearn.metrics import mean_absolute_percentage_error
    print("maPe-Testset:",mean_absolute_percentage_error(y_Test,predicted_stock_price_test_ori))
    #print("mape-Trainset:",mean_absolute_percentage_error(y_Train,predicted_stock_price_train_ori))
#rmsemape that calculates two performance metrics for a time series prediction model: 
    #Root Mean Squared Error (RMSE) and Mean Absolute Percentage Error (MAPE).
```

```python
from sklearn.metrics import mean_squared_error
trends=['n','t','c','ct']
lags=[1,2,3]
from statsmodels.tsa.ar_model import AutoReg
for i in lags:
    #print(i)
   
    for td in trends:
        
        #print(td)
        #trendslist.append(td)
        model = AutoReg(X_train, lags=i,trend=td)
        model_fit = model.fit()
        # make prediction
        y_pred= model_fit.predict(len(X_train), len(data1)-1)
        #print(y_pred)
        #mse=mean_squared_error(y_test,y_pred,squared=False)
        from stockFunctions import rmsemape
        print("Lag={},Trend={}".format(i,td))
        rmsemape(y_test,y_pred)  #Root Mean Squared Error (RMSE) and Mean Absolute Percentage Error (MAPE) of the predictions.
        print("************")
```
